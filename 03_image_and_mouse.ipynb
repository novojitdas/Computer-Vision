```python
### LEFT & RIGHT MOUSE BUTTON EVENT ###

import cv2
import numpy as np 

################
### FUNCTION ###
################

#Callback Function
def draw_circle(event,x,y,flags,param): #parameters will be filled by setMouseCallback
     
        if event == cv2.EVENT_LBUTTONDOWN:
            cv2.circle(img,(x,y),radius=10,color=(255,0,0),thickness=-1) #COLOR: BGR
        elif event == cv2.EVENT_RBUTTONDOWN:
            cv2.circle(img,(x,y),radius=10,color=(0,0,255),thickness=-1)

# CONNECT THE IMAGE with namedWindow.
# The function namedWindow creates a window that can be used as- 
# -a placeholder for images and trackbars.
cv2.namedWindow(winname='my_drawing')

# Set the Mouse Callback   
cv2.setMouseCallback('my_drawing',draw_circle) # (Image WindowName,Callback Function)



##################################
### Showing Images With OpenCV ###
##################################

img = np.zeros((512,512,3))

while True:
    
    #show window & image
    cv2.imshow('my_drawing',img) #my_drawing (don't put any spaces on window name) 
    
    ##If We have waited 20ms & Pressed the Esc
    if cv2.waitKey(20) & 0xFF == 27:
        break 
        
cv2.destroyAllWindows()
```

```python
### DRAW RACTANGLE WITH MOUSE ### 

import cv2 
import numpy as np

###############
## VARIABLES ##
###############
# true while mouse button down,false while mouse button up.
drawing = False
ix = -1
iy = -1


##############
## FUNCTION ##
##############
def draw_rectangle(event,x,y,flags,params): #parameters will be filled by setMouseCallback
    
    global drawing,ix,iy
    
    if event == cv2.EVENT_LBUTTONDOWN:
        
        drawing = True
        ix,iy = x,y  
        
    elif event == cv2.EVENT_MOUSEMOVE:
        if drawing == True:
            cv2.rectangle(img,(ix,iy),(x,y),(0,255,0),-1) # starting (ix,iy) ending (x,y)
        
    elif event == cv2.EVENT_LBUTTONUP:
        drawing = False
        cv2.rectangle(img,(ix,iy),(x,y),(0,255,0),-1)
        
        
#######################           
## SHOWING THE IMAGE ##
#######################
# black
img = np.zeros((512,512,3))

# The function namedWindow creates a window that can be used as- 
# -a placeholder for images and trackbars.
cv2.namedWindow(winname='my_drawing')

cv2.setMouseCallback('my_drawing',draw_rectangle)  # (Image WindowName,Callback Function)

while True:
    
    #show window & image
    cv2.imshow('my_drawing',img) #my_drawing (don't put any spaces on window name)
    
    if cv2.waitKey(1) & 0xFF == 27:
        break 

cv2.destroyAllWindows()
```
