### Import Image Using OpenCV

```python
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# import opencv library 
import cv2
```

```python
# image reading function of open cv
# it directly store the image into numpy array
img = cv2.imread('DATA/00-puppy.jpg')
```

```python
# checking image array type
type(img)
```

```python
# checking the shape 
img.shape
```

```python
#plotting image as open cv default color map
plt.imshow(img)
```

### FIX Image


```python
#MATPLOTLIB ---> RGB ( RED GREEN BLUE)
# OPEN CV   ---> BGR ( BLUE GREEN RED)
```

```python
# Now converting color channel from br 
fix_img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(fix_img)
```

### GrayScale Image Importing

```python
# read image and import as grayscale
gray_img = cv2.imread('DATA/00-puppy.jpg',cv2.IMREAD_GRAYSCALE)
```

```python
# Default color mapping of imshow 
# shows the darker value as blue 
# shows the lighter value as yellowish green
plt.imshow(gray_img)
```

```python
# cv2.IMREAD_GRAYSCALE is imported this image as only one channel
gray_img.shape
```

```python
# now changing color mapping to proper gray 
plt.imshow(gray_img,cmap='gray')
```

### Resize

```python
plt.imshow(fix_img)
```

```python
fix_img.shape
# OUTPUT -> (X,Y,Channels)
```

```python
resized_img = cv2.resize(fix_img,(1000,400))
# cv2.resize(img,(Y,X))
```

```python
plt.imshow(resized_img)
```

```python
weight_ratio = 0.1 
height_ratio = 0.1 # 10% of original height
```

```python
mini_img = cv2.resize(fix_img,(0,0),fix_img,weight_ratio,height_ratio)
plt.imshow(mini_img)
```

### FLIP Image

```python
# Filpped to upside down
flip_img = cv2.flip(fix_img,0)
plt.imshow(flip_img)
```

```python
# Mirror rotate
flip_img = cv2.flip(fix_img,1)
plt.imshow(flip_img)
```

```python
# For Both -1
flip_img = cv2.flip(fix_img,-1)
plt.imshow(flip_img)
```

### Write Image

```python
type(flip_img)
```

```python
# (Destination Path,Source Variable Name)
cv2.imwrite('DATA/01-fliped-puppy.jpg',flip_img)
```
