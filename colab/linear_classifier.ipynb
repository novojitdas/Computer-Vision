#Parameterized Learning with Linear Image Classifier

**KNN** Can't function without maintaining a replica of data inside. If Training set is only few megabytes, then it won't create any problem. But In the Large Dataset lije ImageNet, where contained 14 Millions Annotated Images. It will cost more storage cost and network overhead.

Instead a more desirable approach is using machine learning model that can learn pattern from input data during training time. small number of parameters that can easily be used to represent the model. this type of machine learning is called **Parameterized learning**.  

Parameterized Classification algorithms learns pattterns from their training data. Parameterization is the process of defining the necessary parameters of a given model.

- **Data**<br>This Component is our input data that we are going to learn from. This data includes both the data points and labels.
- **Scoring Function**<br>
For an input image, some function f(scoring function) will applies and then returns the predicted class.

- **Loss Function**<br>
A loss function tells how well our prediction works. lower the loss and higher classification accuracy could gained on at least on the training set.

- **Weights and biases**<br>
The Weight Matrix ***W*** and bias vector **b** will be tweaking and fiddling to increase classification accuracy. Optimization methods that allow us to find a set of parameters ***W*** and **b** that minimize our loss function.

**Scoring Function =** ***W*** . Xi + b (Where W is weight matrix, b bias vector and Xi is image. )


#Advantages of Parameterized learning and linear classification

1. Once we are done training our model, we can discard the input data and keep only the weight matrix W and the bias vector b.

2. Classifying new test data is fast.

now creating simple lnear classifier.

```python
import numpy as np #for numerical processing
import cv2 #opencv to load our image
import matplotlib.pyplot as plt
%matplotlib inline

# class labels
labels = ["dog","cat","panda"]
#random seed
np.random.seed(1)
W = np.random.randn(3,3072) # (3 rows, 3072 cols)
b = np.random.randn(3) # bias vector
input_image = cv2.imread("/content/dog.jpg")
image = cv2.resize(input_image,(32,32)).flatten()
scores = W.dot(image) + b

#loop iver the scores + labels and display them
for (label,score) in zip(labels,scores):
  print("[INFO] {}: {:.2f}".format(label,score))


#printing image
plt.imshow(input_image)
```
